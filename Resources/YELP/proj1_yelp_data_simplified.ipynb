# Extracting Data from Yelp Buisness (Simplified)

## Install Dependencies

```python
import pandas as pd
import json 
import csv
import numpy as np
from tqdm import tqdm
```

```python
# Link to the json 
businesses_link = '/Resources/YELP/yelp_academic_dataset_business.json'
```

```python
# Load json
with open(businesses_link, 'r') as f:
    businesses = [json.loads(line) for line in f]
```

```python
# read into dataframe
biz_data = pd.DataFrame(businesses)
```

```python
biz_data.describe()
```

```python
biz_data.head()
```

### Locating all locations in Philadelphia

```python
filter1 = biz_data[biz_data['city'].isin(["Philadelphia"])]
```

```python
filter1
```

## Exporting into CSV

```python
# export the dataframe to a CSV file with high efficiency
chunk_size = 10000 # adjust this to a size that works for your system
df_list = [chunk for _, chunk in filter1.groupby(np.arange(len(filter1)) // chunk_size)]

with open('../yelp_locations_df_v01.csv', 'w') as f:
    writer = csv.writer(f)
    pbar = tqdm(total=len(filter1))
    for chunk in df_list:
        chunk.to_csv(f, index=False, header=False)
        pbar.update(len(chunk))
    pbar.close()
```
