```python
import csv
import pandas as pd
import numpy as np
import scipy
from tqdm import tqdm
```

### Spotify

```python
# Defining a generator to save memory on reading csv
spotify_raw = "./proj1_spotify_data_v0.csv"
chunksizel = 1000

def prius_csv():
    reader = pd.read_csv(spotify_raw, chunksize=chunksizel)
    chunks = []
    for chunk in tqdm(reader):
        chunks.append(chunk)
    df = pd.concat(chunks)
    return df

spotify_processed = prius_csv()
```

#### NFL Schedules

```python
# Defining a generator to save memory on reading csv
spotify_raw = "./seasons/regular_season.csv"

def camry_csv():
    reader = pd.read_csv(spotify_raw, chunksize=chunksizel)
    chunks = []
    for chunk in tqdm(reader):
        chunks.append(chunk)
    df = pd.concat(chunks)
    return df

nfl_schedules = camry_csv()
```

```python
nfl_schedules = nfl_schedules.dropna(subset=['game loc'])
nfl_schedules
```

### Merging Datasets

```python
# Filter by 'year' and 'month'
merged1_df = pd.merge(spotify_processed, nfl_schedules, on=['year', 'month'], how='inner')
```

```python
# Make dataset smaller
merged1_df.drop('Unnamed: 0_x', axis=1, inplace=True)
merged1_df.drop('track_id', axis=1, inplace=True)
merged1_df.drop('combined week date', axis=1, inplace=True)
merged1_df.drop('day_x',axis=1,inplace=True)
```

```python
#Make dataset smaller
merged1_df.drop('Unnamed: 0_y', axis=1, inplace=True)
merged1_df.drop('day_y', axis=1, inplace=True)
merged1_df.drop('score', axis=1, inplace=True)
merged1_df.drop('year', axis=1, inplace=True)
merged1_df.drop('month', axis=1, inplace=True)
```

#### Generate subsets for analysis

```python
# Music Duration Dataframe 

MbD_df = merged1_df

# Create a new column to categorize duration as short, radio-ready, or long
MbD_df['duration_cat'] = pd.cut(MbD_df['duration'], bins=[0, 120000, 240000, float('inf')], labels=['short', 'radio-ready', 'long'])
```

```python
# Create a new column to categorize dates as quarters
MbD_df['quarter'] = pd.PeriodIndex(MbD_df['date'], freq='Q')
```

```python
# Bug fix to allow special characters and be able to group by artist
MbD_df['artists'] = MbD_df['artists'].str.replace('$', r'\$', regex=True)
```

```python
MbD_df.head()
```

##### Distribution of Songs by Duration Category by Year

```python
# Subset needed for charting, group by 
MbD_chart2 = MbD_df.loc[:,['name','streams','date','quarter','duration_cat','duration','week number']]
```

```python
# Defining the terms of grouping 
MbD_df_chart2_grp = MbD_chart2.groupby('name').agg({\
    'streams': ['mean', 'sum','count'], 
    'duration': ['mean', 'sum'],
    'quarter': ['first','last'],
    })
```

```python
# Formatting
MbD_df_chart2_grp.columns = [f"{col[0]}_{col[1].lower()}" for col in MbD_df_chart2_grp.columns]
MbD_df_chart2_grp = MbD_df_chart2_grp.reset_index()
```

```python
# More Formatting
MbD_df_chart2_grp.columns = ['Song Name', 'Mean Streams', 'Total Streams', 'Stream Count', 'Mean Duration', 'Total Duration', 'First Quarter', 'Last Quarter']
MbD_df_chart2_grp
```

##### Distribution of Songs by Streams and Length

```python
# Subset needed for charting, group by NFL week number
MbD_df_chart1 = MbD_df.loc[:,['streams','date','quarter','duration_cat','duration','week number']]
```

```python
# Defining the terms of grouping 
MbD_df_chart1_grp = MbD_df_chart1.groupby('week number').agg({\
    'streams': ['mean', 'sum','count'], 
    'duration': ['mean', 'sum'],
    'quarter': ['first','last'],
    })
```

```python
# Formatting
MbD_df_chart1_grp.columns = [f"{col[0]}_{col[1].lower()}" for col in MbD_df_chart1_grp.columns]
MbD_df_chart1_grp = MbD_df_chart1_grp.reset_index()
MbD_df_chart1_grp.columns = ['Week Number', 'Mean Streams', 'Total Streams','Stream Count', 'Mean Duration', 'Total Duration', 'First Quarter', 'Last Quarter']
```
