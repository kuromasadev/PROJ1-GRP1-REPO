## Extracting Spotify Data

### Importing Dependencies

```python
import csv
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
# Defining a generator to save memory on reading csv
spotify_raw = "../spotify/charts.csv"
chunksizel = 1000

def prius_csv():
    reader = pd.read_csv(spotify_raw, chunksize=chunksizel)
    chunks = []
    for chunk in tqdm(reader):
        chunks.append(chunk)
    df = pd.concat(chunks)
    return df

spotify_processed = prius_csv()
```

## Get durations from the NFL schedules

#### Obtain Schedules for given years

```python
from bs4 import BeautifulSoup
import requests
import os
```

```python
# establish variables
nfl_base_url = "https://www.jt-sw.com/football/pro/results.nsf/Teams/"
nfl_seasons = [2015,2016,2017,2018,2019,2020,2021]
nfl_team = "-phi"

# create a seasons subfolder if it doesn't exist
if not os.path.exists('seasons'):
    os.makedirs('seasons')

# Loop through each year and export each subtable as a dataframe
for season in nfl_seasons:
    nfl_url = nfl_base_url + str(season) + nfl_team

    # get response 
    response = requests.get(nfl_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Use Beautiful Soup to find all the tables within the "table" tag of the HTML content
    table_tag = soup.find('table')
    subtable_tag = table_tag.find_all('table')

    # Loop through each subtable and convert to dataframe
    for i, subtable in enumerate(subtable_tag):
        df = pd.read_html(str(subtable))[0]
        df.insert(0, 'year', season)  # Add year column
        df.drop(df.columns[2], axis=1, inplace=True)  # Drop index 1 column
        # Normal full seasons will be identified by 0
        # Special appearances in either playoffs, wildcards, conference championships or superbowls
        df.to_csv(f"seasons/nfl_{nfl_team.strip('-')}_season_{season}_table_{i}.csv", index=False)
```

### Reload Schedules into schedule dataframe

```python
regular_season_files = ['./seasons/nfl_phi_season_2015_table_0.csv',
                        './seasons/nfl_phi_season_2016_table_0.csv',
                        './seasons/nfl_phi_season_2017_table_0.csv',
                        './seasons/nfl_phi_season_2018_table_0.csv',
                        './seasons/nfl_phi_season_2019_table_0.csv',
                        './seasons/nfl_phi_season_2020_table_0.csv',
                        './seasons/nfl_phi_season_2021_table_0.csv'
                        ]
```

```python
regular_seasons_df = []
```

```python
# Loop through the list of file paths and read each CSV file into a dataframe

for file in regular_season_files:
    rs_df = pd.read_csv(file)
    regular_seasons_df.append(rs_df)
```

```python
# Concatenate all dataframes into a single dataframe
nfl_phi_seasons = pd.concat(regular_seasons_df, ignore_index=True)
```

```python
# Rename the columns from numbers to actual column names
nfl_phi_seasons = nfl_phi_seasons.rename(columns={\
    '0':'week number',
    '2':'combined week date',
    '3':'result',
    '4':'score',
    '5':'game loc'})
```

```python
# convert game loc column to string
nfl_phi_seasons['game loc'] = nfl_phi_seasons['game loc'].astype(str)

# define a lambda function to implement the logic for the new column
turf_adv = lambda x: "away" if x.startswith("at") else ("home" if x.startswith("vs") else "no game")

# apply the lambda function to the 'game loc' column to create the new column
nfl_phi_seasons["turf adv"] = nfl_phi_seasons["game loc"].apply(turf_adv)
```

```python
# Split combined week date into month number and day in separate columns
date_df = nfl_phi_seasons['combined week date'].str.strip().str.split(' ', expand=True)
date_df.columns = ['month', 'day']

# Define dictionary of month names to numbers
month_dict = {'Jan': '1',
              'Feb': '2',
              'Mar': '3',
              'Apr': '4',
              'May': '5',
              'Jun': '6',
              'Jul': '7',
              'Aug': '8',
              'Sep': '9',
              'Oct': '10',
              'Nov': '11',
              'Dec': '12'}

# Apply dictionary to month column
date_df['month'] = date_df['month'].map(month_dict)
```

```python
nfl_phi_seasons = pd.concat([nfl_phi_seasons, date_df], axis=1)
```

```python
# Export season calendar for later analysis
nfl_phi_seasons.to_csv("./seasons/regular_season.csv")
```

## Use NFL Schedule as filter for music

```python
# Extract only the columns needed to filter spotify data
schedule_filter = nfl_phi_seasons[['week number','year','month','day']]
```

```python
# Filter out bye weeks using boolean indexing
schedule_filter = schedule_filter[~schedule_filter['day'].str.contains('Week')]
```
